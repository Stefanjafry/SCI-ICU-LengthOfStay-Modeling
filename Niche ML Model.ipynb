## Introduction :

this section introduces a niche-based modeling strategy for predicting ICU Length of Stay (LOS) in spinal cord injury (SCI) patients. The EDA revealed significant heterogeneity across patient subgroups—particularly by injury type, admission source, and surgical vs. non-surgical pathway—which suggests that a single, monolithic model may fail to capture subgroup-specific patterns in LOS dynamics.

To address this, we segment the population into clinically meaningful niches using a combination of:

Injury classification (e.g., cervical vs. thoracic injuries)

Admission type (e.g., elective vs. emergency)

Discharge disposition and mortality status

Dimensionality-reduced cluster assignments (e.g., PCA + k-means or hierarchical clustering)

Each niche group is treated as a distinct sub-cohort, and a dedicated regression model is trained per group to better capture localized patterns, non-linearities, and interaction effects that may be obscured in a global model. This approach offers several benefits:

Improved predictive accuracy by reducing intra-group variance

Enhanced model interpretability for clinicians within specific clinical contexts

A modular architecture that can be expanded as new cohorts or injury categories emerge

The models are evaluated using standard regression metrics (RMSE, MAE, R²), with special attention paid to performance consistency across niches and the extent of overfitting in small subgroups. This niche modeling framework lays the groundwork for precision resource allocation and personalized ICU planning in SCI care.

### Preprocessing df:

```python
# Define MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# --- Preprocessing ---
df.drop(columns=['admission_type', 'admission_location'], inplace=True, errors='ignore')
df['marital_status'].fillna('Unknown')

# --- Categorical Encoding ---
expected_categorical_cols = [
    'gender', 'insurance', 'marital_status',
    'long_title_grouped', 'race_grouped',
    'admission_type_grouped', 'admission_location_grouped'
]
categorical_cols = [col for col in expected_categorical_cols if col in df.columns]

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

### Cluster 0(Referral Patients):

```python
# --- Filter for Cluster 0 ---
df_cluster_0 = df[df['cluster'] == 0].copy()
target_0 = 'los'
groups_0 = df_cluster_0['subject_id']
X_0 = df_cluster_0.drop(columns=['subject_id', target_0])
y_0 = df_cluster_0[target_0]

# --- Initialize Metrics and CV ---
rmse_scores_0, mae_scores_0, r2_scores_0, mape_scores_0, medae_scores_0 = [], [], [], [], []
gkf_0 = GroupKFold(n_splits=5)

y_true_all_lgb_0, y_pred_all_lgb_0 = [], []

# --- Cross-validation Loop ---
for fold, (train_idx_0, val_idx_0) in enumerate(gkf_0.split(X_0, y_0, groups_0)):
    print(f"\nFold {fold + 1}/5")

    X_train_0, X_val_0 = X_0.iloc[train_idx_0], X_0.iloc[val_idx_0]
    y_train_0, y_val_0 = y_0.iloc[train_idx_0], y_0.iloc[val_idx_0]

    model_0 = LGBMRegressor(
        objective='regression',
        learning_rate=0.05,
        n_estimators=1000
    )

    model_0.fit(
        X_train_0, y_train_0,
        eval_set=[(X_val_0, y_val_0)],
        eval_metric='rmse',
        categorical_feature=categorical_cols,
        callbacks=[
            early_stopping(stopping_rounds=20),
            log_evaluation(period=50)
        ]
    )

    y_pred_0 = model_0.predict(X_val_0)
    y_true_all_lgb_0.extend(y_val_0)
    y_pred_all_lgb_0.extend(y_pred_0)

    # --- Evaluation ---
    rmse_0 = np.sqrt(mean_squared_error(y_val_0, y_pred_0))
    mae_0 = mean_absolute_error(y_val_0, y_pred_0)
    medae_0 = median_absolute_error(y_val_0, y_pred_0)
    r2_0 = r2_score(y_val_0, y_pred_0)
    mape_0 = mean_absolute_percentage_error(y_val_0, y_pred_0)

    print(f"  RMSE : {rmse_0:.4f}")
    print(f"  MAE  : {mae_0:.4f}")
    print(f"  MedAE: {medae_0:.4f}")
    print(f"  R²   : {r2_0:.4f}")
    print(f"  MAPE : {mape_0:.2f}%")

    # Store metrics
    rmse_scores_0.append(rmse_0)
    mae_scores_0.append(mae_0)
    medae_scores_0.append(medae_0)
    r2_scores_0.append(r2_0)
    mape_scores_0.append(mape_0)

# --- Final Summary ---
print("\n====== Cluster 0 Cross-Validation Summary ======")
print(f"Average RMSE : {np.mean(rmse_scores_0):.4f}")
print(f"Average MAE  : {np.mean(mae_scores_0):.4f}")
print(f"Average MedAE: {np.mean(medae_scores_0):.4f}")
print(f"Average R²   : {np.mean(r2_scores_0):.4f}")
print(f"Average MAPE : {np.mean(mape_scores_0):.2f}%")
```
