## Introduction:

building predictive models aimed at estimating ICU Length of Stay (LOS) for patients with spinal cord injuries. Given the substantial clinical and administrative complexity observed in the EDA phase—such as skewed LOS distributions, class imbalances, and mortality-related confounding—traditional linear modeling approaches are likely to be inadequate.

This phase will explore supervised regression models that can handle:

Non-linear feature interactions

Heterogeneous patient subgroups

Outlier robustness

Censored or truncated outcomes due to in-hospital mortality

To ensure generalizable performance and prevent data leakage, we will use GroupKFold cross-validation, grouping by subject_id to account for repeated admissions or correlated records. Evaluation metrics such as RMSE, MAE, MedAE, R², and MAPE will be employed to capture both central tendency and distributional accuracy.

Model selection, feature engineering, and calibration techniques will be guided by the statistical and clinical patterns uncovered during EDA. The goal is to develop models that not only achieve good predictive performance but also maintain clinical interpretability and fairness across patient subgroups.

Next, Data cleaning needs to be conducted. WBC has <1% coverage, making it an extremely sparse feature. Therefore, we will drop these rows. Next, numerical features are overall extremely clean, with 2-36 rows missing in columns, so inverse transformation was used to impute the data. We used inverse transformation because some numerical columns did not exhibit a normal distribution, so imputing the Mean, Median and Mode would introduce bias. Furthermore, the date of death will be turned into a binary indicator where 1=death and 0=survived so the model can properly interpret this feature. Next, extremely rare injuries that only happen to 1 or 2 patients will be aggregated into a rare injury column. This is done because the plots were practically horizontal lines and extremely high in our EDA (the box plot shows the bottom 10 injury types, with LOS as the Y-axis). Although not perfect (having more data points would be better as 1 or 2 data points leave no room for data synthisization), we will aggregate them into a rare column to reduce overfitting to those variables. The race columns is extremely specific, so they will be aggregated as well using the formula in the column below. Admission type and admission location will be consolidated into broader more medically meaningful categories. Last leaky features such as outtime will be removed.

White = WHITE, WHITE - OTHER EUROPEAN, WHITE - RUSSIAN, WHITE - BRAZILIAN, WHITE - EASTERN EUROPEAN
Black = BLACK/AFRICAN AMERICAN, BLACK/CAPE VERDEAN, BLACK/CARIBBEAN ISLAND, BLACK/AFRICAN
Latino = HISPANIC OR LATINO, HISPANIC/LATINO - PUERTO RICAN, DOMINICAN, GUATEMALAN, CENTRAL AMERICAN, SALVADORAN, HONDURAN, CUBAN, COLOMBIAN, MEXICAN
Asian = ASIAN, ASIAN - CHINESE, ASIAN - ASIAN INDIAN, ASIAN - SOUTH EAST ASIAN, ASIAN - KOREAN
Native american = MERICAN INDIAN/ALASKA NATIVE, NATIVE HAWAIIAN OR OTHER PACIFIC ISLANDER
Other = OTHER, PORTUGUESE, SOUTH AMERICAN, MULTIPLE RACE/ETHNICITY
unkown = UNKNOWN, UNABLE TO OBTAIN, PATIENT DECLINED TO ANSWER 

```python
# Define WBC-related columns to drop
wbc_columns = ['avg_wbc', 'min_wbc', 'max_wbc']
# Drop the columns from the DataFrame
df = df.drop(columns=wbc_columns)

# Define inverse transform sampling function
def inverse_transform_impute(column):
    observed = column.dropna().values
    n_missing = column.isna().sum()
    # Sample from observed distribution
    imputed = np.random.choice(observed, size=n_missing, replace=True)
    # Fill missing entries
    col_filled = column.copy()
    col_filled[column.isna()] = imputed
    return col_filled

#Only check numeric columns for missingness
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
num_cols_with_nan = [col for col in numeric_cols if df[col].isna().any()]
# Apply inverse transform imputation
for col in num_cols_with_nan:
    df[col] = inverse_transform_impute(df[col])
```

We are using CatBoost and LightGBM on this dataset because they are highly optimized gradient-boosting libraries that deliver state-of-the-art performance on structured, tabular data—precisely the kind of data we have in this spinal injury dataset. We used GroupKFold cross-validation because each patient (subject_id) can appear multiple times in the dataset due to multiple injuries. Since we are predicting LOS some of the most relevant and interpretable metrics we will use are MAE (Interpretable in actual days), RMSE (Captures variance in errors), R^2 (Useful for overall model fit), MAPE (Error as a percentage of actual LOS) and Median Absolute Error (less sensitive to outliers as ICU LOS often has outliers).

```python
# Define MAPE (not built-in)
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# --- Load Data ---
df = pd.read_csv("cleaned_spinal_patient_features.csv")

# --- Preprocessing ---
df.drop(columns=['admission_type', 'admission_location'], inplace=True, errors='ignore')
df['marital_status'].fillna('Unknown', inplace=True)

# --- Categorical Encoding ---
expected_categorical_cols = [
    'gender', 'insurance', 'marital_status',
    'long_title_grouped', 'race_grouped',
    'admission_type_grouped', 'admission_location_grouped'
]
categorical_cols = [col for col in expected_categorical_cols if col in df.columns]

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# --- Prepare Features and Target ---
target = 'los'
groups = df['subject_id']
X = df.drop(columns=['subject_id', target])
y = df[target]

# --- Initialize Metrics and Cross-Validation ---
rmse_scores, mae_scores, r2_scores, mape_scores, medae_scores = [], [], [], [], []
gkf = GroupKFold(n_splits=5)

# --- Cross-Validation Loop ---
y_true_all_lgb, y_pred_all_lgb = [], []
for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y, groups)):
    print(f"\nFold {fold + 1}/5")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = LGBMRegressor(
        objective='regression',
        learning_rate=0.05,
        n_estimators=1000
    )

    model.fit(
        X_train, y_train,
        eval_set=[(X_val, y_val)],
        eval_metric='rmse',
        categorical_feature=categorical_cols,
        callbacks=[
            early_stopping(stopping_rounds=20),
            log_evaluation(period=50)
        ]
    )

    y_pred = model.predict(X_val)
    y_true_all_lgb.extend(y_val)
    y_pred_all_lgb.extend(y_pred)

    # --- Evaluation ---
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    mae = mean_absolute_error(y_val, y_pred)
    medae = median_absolute_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    mape = mean_absolute_percentage_error(y_val, y_pred)

    print(f"Fold {fold + 1} Metrics:")
    print(f"  RMSE : {rmse:.4f}")
    print(f"  MAE  : {mae:.4f}")
    print(f"  MedAE: {medae:.4f}")
    print(f"  R²   : {r2:.4f}")
    print(f"  MAPE : {mape:.2f}%")

    # Store metrics
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    medae_scores.append(medae)
    r2_scores.append(r2)
    mape_scores.append(mape)

y_true_all_lgb = np.array(y_true_all_lgb)
y_pred_all_lgb = np.array(y_pred_all_lgb)

# --- Final Summary ---
print("\n====== Cross-Validation Summary ======")
print(f"Average RMSE : {np.mean(rmse_scores):.4f}")
print(f"Average MAE  : {np.mean(mae_scores):.4f}")
print(f"Average MedAE: {np.mean(medae_scores):.4f}")
print(f"Average R²   : {np.mean(r2_scores):.4f}")
print(f"Average MAPE : {np.mean(mape_scores):.2f}%")
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import GroupKFold
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error,
    median_absolute_error, r2_score
)

# Define MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# --- Load & Preprocess ---
df = pd.read_csv("cleaned_spinal_patient_features.csv")
df.drop(columns=['admission_type', 'admission_location'], inplace=True, errors='ignore')
df['marital_status'].fillna('Unknown', inplace=True)

# Categorical feature handling
expected_categorical_cols = [
    'gender', 'insurance', 'marital_status',
    'long_title_grouped', 'race_grouped',
    'admission_type_grouped', 'admission_location_grouped'
]
categorical_cols = [col for col in expected_categorical_cols if col in df.columns]

# Feature and target
target = 'los'
groups = df['subject_id']
X = df.drop(columns=['subject_id', target])
y = df[target]

# Metrics and residual arrays
rmse_scores, mae_scores, medae_scores, r2_scores, mape_scores = [], [], [], [], []
y_true_all_cb, y_pred_all_cb = [], []

# --- GroupKFold Cross-Validation ---
gkf = GroupKFold(n_splits=5)
for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y, groups)):
    print(f"\nFold {fold + 1}/5")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # CatBoost Pool
    train_pool = Pool(data=X_train, label=y_train, cat_features=categorical_cols)
    val_pool = Pool(data=X_val, label=y_val, cat_features=categorical_cols)

    # Initialize & train
    model = CatBoostRegressor(
        iterations=1000,
        learning_rate=0.05,
        depth=6,
        loss_function='RMSE',
        eval_metric='RMSE',
        random_seed=42,
        early_stopping_rounds=20,
        verbose=50
    )
    model.fit(train_pool, eval_set=val_pool)

    # Predict & collect
    y_pred = model.predict(X_val)
    y_true_all_cb.extend(y_val)
    y_pred_all_cb.extend(y_pred)

    # Metrics
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    mae = mean_absolute_error(y_val, y_pred)
    medae = median_absolute_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    mape = mean_absolute_percentage_error(y_val, y_pred)

    print(f"Fold {fold + 1} Metrics:")
    print(f"  RMSE : {rmse:.4f}")
    print(f"  MAE  : {mae:.4f}")
    print(f"  MedAE: {medae:.4f}")
    print(f"  R²   : {r2:.4f}")
    print(f"  MAPE : {mape:.2f}%")

    # Store
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    medae_scores.append(medae)
    r2_scores.append(r2)
    mape_scores.append(mape)

# Convert predictions for analysis
y_true_all_cb = np.array(y_true_all_cb)
y_pred_all_cb = np.array(y_pred_all_cb)

# --- Final Summary ---
print("\n====== Cross-Validation Summary ======")
print(f"Average RMSE : {np.mean(rmse_scores):.4f}")
print(f"Average MAE  : {np.mean(mae_scores):.4f}")
print(f"Average MedAE: {np.mean(medae_scores):.4f}")
print(f"Average R²   : {np.mean(r2_scores):.4f}")
print(f"Average MAPE : {np.mean(mape_scores):.2f}%")
```

Independently, LightGBM achieved an RMSE of 3.95, MAE of 2.01, MedAE of 1.01, and R² of 0.556, showing moderate prediction ability with a skew towards generalizing relatively well to patients. It had a larger MAPE of 89.5%, indicating higher relative errors for shorter LOS cases. CatBoost, on the other hand, achieved a lower RMSE (3.91) and MAPE (83.9%), a slightly higher R² (0.565), and almost identical MAE and MedAE, indicating better performance and better calibration, especially on short LOS cases. Overall, as of now, CatBoost is slightly outperforming LightGBM

We will now move on to residual analysis. We will create QQ plots which is used to helps evaluate the normality of residuals, deviations from the 45-degree line indicate skewness or kurtosis issues. The fitted vs. residuals plot checks the assumption of homoscedasticity (constant variance of errors) and helps identify non-linearity; ideally, residuals should scatter randomly around zero with no clear pattern. Studentized residuals helps assess the normality assumption in regression by showing whether residuals approximate a bell-shaped, symmetric distribution centered at zero. Deviations such as skewness or heavy tails may indicate outliers, non-linearity, or model misspecification.

```python
# Compute residuals
residuals_lgb = y_true_all_lgb - y_pred_all_lgb
residuals_cb = y_true_all_cb - y_pred_all_cb

# Studentized residuals function
def studentized_residuals(y_true, y_pred):
    residuals = y_true - y_pred
    mse = np.mean(residuals ** 2)
    leverage = 1 / len(y_true)  # approximate, no hat matrix
    return residuals / np.sqrt(mse * (1 - leverage))

# Compute studentized residuals
stud_resid_lgb = studentized_residuals(y_true_all_lgb, y_pred_all_lgb)
stud_resid_cb = studentized_residuals(y_true_all_cb, y_pred_all_cb)

# Plot diagnostics
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# LightGBM QQ plot
stats.probplot(stud_resid_lgb, dist="norm", plot=axes[0, 0])
axes[0, 0].set_title("LightGBM QQ Plot")

# LightGBM Residuals vs Fitted
axes[0, 1].scatter(y_pred_all_lgb, residuals_lgb, alpha=0.5)
axes[0, 1].axhline(0, color='red', linestyle='--')
axes[0, 1].set_title("LightGBM Residuals vs Fitted")
axes[0, 1].set_xlabel("Fitted Values")
axes[0, 1].set_ylabel("Residuals")

# LightGBM Studentized Residuals
axes[0, 2].hist(stud_resid_lgb, bins=30, edgecolor='k')
axes[0, 2].set_title("LightGBM Studentized Residuals")

# CatBoost QQ plot
stats.probplot(stud_resid_cb, dist="norm", plot=axes[1, 0])
axes[1, 0].set_title("CatBoost QQ Plot")

# CatBoost Residuals vs Fitted
axes[1, 1].scatter(y_pred_all_cb, residuals_cb, alpha=0.5)
axes[1, 1].axhline(0, color='red', linestyle='--')
axes[1, 1].set_title("CatBoost Residuals vs Fitted")
axes[1, 1].set_xlabel("Fitted Values")
axes[1, 1].set_ylabel("Residuals")

# CatBoost Studentized Residuals
axes[1, 2].hist(stud_resid_cb, bins=30, edgecolor='k')
axes[1, 2].set_title("CatBoost Studentized Residuals")

# Compute residuals
df['residual_lgb'] = y_true_all_lgb - y_pred_all_lgb
df['residual_cb'] = y_true_all_cb - y_pred_all_cb

# Plot residual histograms
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].hist(residuals_lgb, bins=30, edgecolor='k')
ax[0].set_title("LightGBM Residual Histogram")
ax[0].set_xlabel("Residual")
ax[0].set_ylabel("Count")

ax[1].hist(residuals_cb, bins=30, edgecolor='k')
ax[1].set_title("CatBoost Residual Histogram")
ax[1].set_xlabel("Residual")
ax[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

Both QQ plots show significant deviations in the tails, indicating non-normality and outliers. For our fitted vs. residuals graph, both models demonstrate a clear funnel pattern indicating heteroscedasticity for residual vs. fitted values. The model's errors are not constant and worsen as LOS increases. Most studentized residuals cluster around 0; however, both models show extreme outliers, implicating extreme prediction errors that can negatively influence features like RMSE. For our residual histogram, several under and over-predictions demonstrate how the model misfits at both extremes.

```python
# Define MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# --- Load Data ---
df = pd.read_csv("cleaned_spinal_patient_features.csv")
df.drop(columns=['admission_type', 'admission_location'], inplace=True, errors='ignore')
df['marital_status'].fillna('Unknown', inplace=True)

# --- Label Encode Categoricals ---
expected_categorical_cols = [
    'gender', 'insurance', 'marital_status',
    'long_title_grouped', 'race_grouped',
    'admission_type_grouped', 'admission_location_grouped'
]
categorical_cols = [col for col in expected_categorical_cols if col in df.columns]

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# --- Feature Engineering ---
df['age_dxcount'] = df['anchor_age'] * df['dx_count']
df['spo2_rr'] = df['min_SpO2'] * df['max_rr']
df['hr_variability'] = df['max_hr'] - df['min_hr']
df['renal_severity'] = df['max_creatinine'] - df['min_creatinine']
df['elderly_flag'] = (df['anchor_age'] >= 70).astype(int)
df['multi_morbidity'] = (df['dx_count'] >= 3).astype(int)

# --- Prepare for Modeling ---
target = 'los'
groups = df['subject_id']
X = df.drop(columns=['subject_id', target])
y = df[target]
y_log = np.log1p(y)

# --- Initialize Evaluation Storage ---
rmse_scores, mae_scores, r2_scores, mape_scores, medae_scores = [], [], [], [], []
y_true_all, y_pred_all = [], []

# --- GroupKFold Cross-Validation ---
gkf = GroupKFold(n_splits=5)
for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y_log, groups)):
    print(f"\nFold {fold + 1}/5")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val_log = y_log.iloc[train_idx], y_log.iloc[val_idx]
    y_val = y.iloc[val_idx]  # for metric comparison on original scale

    model = LGBMRegressor(
        objective='huber',
        alpha=0.9,  # controls delta threshold in Huber loss
        learning_rate=0.05,
        n_estimators=1000,
        random_state=42
    )

    model.fit(
        X_train, y_train,
        eval_set=[(X_val, y_val_log)],
        eval_metric='rmse',
        categorical_feature=categorical_cols,
        callbacks=[
            early_stopping(stopping_rounds=20),
            log_evaluation(period=50)
        ]
    )

    # Predict and back-transform
    y_pred_log = model.predict(X_val)
    y_pred = np.expm1(y_pred_log)
    
    y_true_all.extend(y_val)
    y_pred_all.extend(y_pred)

    # Metrics
    rmse_scores.append(np.sqrt(mean_squared_error(y_val, y_pred)))
    mae_scores.append(mean_absolute_error(y_val, y_pred))
    medae_scores.append(median_absolute_error(y_val, y_pred))
    r2_scores.append(r2_score(y_val, y_pred))
    mape_scores.append(mean_absolute_percentage_error(y_val, y_pred))

# --- Final Summary ---
print("\n====== Cross-Validation Summary (Log + Huber) ======")
print(f"Average RMSE : {np.mean(rmse_scores):.4f}")
print(f"Average MAE  : {np.mean(mae_scores):.4f}")
print(f"Average MedAE: {np.mean(medae_scores):.4f}")
print(f"Average R²   : {np.mean(r2_scores):.4f}")
print(f"Average MAPE : {np.mean(mape_scores):.2f}%")

# --- Save Residuals for Analysis ---
residuals_df = pd.DataFrame({
    'True_LOS': y_true_all,
    'Predicted_LOS': y_pred_all,
    'Residual': np.array(y_true_all) - np.array(y_pred_all)
})
residuals_df.to_csv("lgb_log_huber_residuals.csv", index=False)

# Optional: Save metrics
metrics_df = pd.DataFrame({
    'Fold': list(range(1, 6)),
    'RMSE': rmse_scores,
    'MAE': mae_scores,
    'MedAE': medae_scores,
    'R2': r2_scores,
    'MAPE': mape_scores
})
metrics_df.to_csv("lgb_log_huber_metrics.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import GroupKFold
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error, 
    median_absolute_error, r2_score
)

# --- Load and preprocess data ---
df = pd.read_csv("cleaned_spinal_patient_features.csv")
df.drop(columns=['admission_type', 'admission_location'], inplace=True, errors='ignore')
df['marital_status'].fillna('Unknown', inplace=True)

# --- Identify categorical columns for CatBoost ---
categorical_cols = [
    'gender', 'insurance', 'marital_status',
    'long_title_grouped', 'race_grouped',
    'admission_type_grouped', 'admission_location_grouped'
]
categorical_cols = [col for col in categorical_cols if col in df.columns]

# --- Feature engineering ---
df['age_dxcount'] = df['anchor_age'] * df['dx_count']
df['spo2_rr'] = df['min_SpO2'] * df['max_rr']
df['hr_variability'] = df['max_hr'] - df['min_hr']
df['renal_severity'] = df['max_creatinine'] - df['min_creatinine']
df['elderly_flag'] = (df['anchor_age'] >= 70).astype(int)
df['multi_morbidity'] = (df['dx_count'] >= 3).astype(int)
df['age_bucket'] = pd.cut(df['anchor_age'], bins=[0, 40, 60, 80, 120], labels=False)
df['dx_count_bucket'] = pd.qcut(df['dx_count'], q=4, labels=False, duplicates='drop')
df['age_bucket_dx'] = df['age_bucket'] * df['dx_count_bucket']
df['comorbidity_risk'] = (df['dx_count'] >= 5).astype(int)
df['elderly_comorbid_flag'] = ((df['anchor_age'] > 70) & (df['dx_count'] >= 3)).astype(int)

# --- Modeling setup ---
target = 'los'
groups = df['subject_id']
X = df.drop(columns=['subject_id', target])
y = df[target]
y_log = np.log1p(y)

# --- Store evaluation results ---
rmse_scores, mae_scores, medae_scores, r2_scores, mape_scores = [], [], [], [], []
y_true_all, y_pred_all = [], []

gkf = GroupKFold(n_splits=5)

# --- GroupKFold cross-validation ---
for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y_log, groups)):
    print(f"\nFold {fold+1}/5")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val_log = y_log.iloc[train_idx], y_log.iloc[val_idx]
    y_val = y.iloc[val_idx]

    train_pool = Pool(X_train, y_train, cat_features=categorical_cols)
    val_pool = Pool(X_val, y_val_log, cat_features=categorical_cols)

    model = CatBoostRegressor(
        loss_function='Huber:delta=1.0',
        learning_rate=0.05,
        iterations=1000,
        random_seed=42,
        verbose=100,
        early_stopping_rounds=20
    )
    model.fit(train_pool, eval_set=val_pool)

    y_pred_log = model.predict(X_val)
    y_pred = np.expm1(y_pred_log)

    y_true_all.extend(y_val)
    y_pred_all.extend(y_pred)

    # Metrics
    rmse_scores.append(np.sqrt(mean_squared_error(y_val, y_pred)))
    mae_scores.append(mean_absolute_error(y_val, y_pred))
    medae_scores.append(median_absolute_error(y_val, y_pred))
    r2_scores.append(r2_score(y_val, y_pred))
    mape_scores.append(np.mean(np.abs((y_val - y_pred) / y_val)) * 100)

# --- Final summary ---
print("\n====== CatBoost (Log + Huber, Native Cat) Summary ======")
print(f"Average RMSE : {np.mean(rmse_scores):.4f}")
print(f"Average MAE  : {np.mean(mae_scores):.4f}")
print(f"Average MedAE: {np.mean(medae_scores):.4f}")
print(f"Average R²   : {np.mean(r2_scores):.4f}")
print(f"Average MAPE : {np.mean(mape_scores):.2f}%")

# --- Save residuals for analysis ---
residuals_df = pd.DataFrame({
    'True_LOS': y_true_all,
    'Predicted_LOS': y_pred_all,
    'Residual': np.array(y_true_all) - np.array(y_pred_all)
})
residuals_df.to_csv("catboost_native_log_huber_residuals.csv", index=False)
```

CatBoost improved in MAPE (26% drop) and in MAE and MedAE, which means the transformations reduced prediction skewness. For lightGBM, after the transformation, all metrics saw improvements, such as MAE, MedAE, and R2. Overall, LightGBM outperformed CatBoost slightly across all metrics after this transformation. The difference is extremely small, however, indicating the equality in performance for these models.

## Conclusion:

In order to systematically fix residuals and model accuracy, clustering will need to be performed to create more niche machine-learning models to help identify extreme outliers (which we are doing). Furthermore, in order to fix our general machine learning model, more SQL will need to be conducted, including variables such as ventilators, discharge policy, surgical information, and bed availability, which would help increase our prediction accuracy and reduce residuals. However, due to the granularity of our project, we ran out of time to perform this as it is an extensive procedure. Additionally, LOS is inherently noisy and hard to predict; even with perfect vitals and labs, non-clinical delays also introduce irreducible variance. The MIMIC IV database is also missing some potential high-impact features, such as insurance approval delays, a substantial economic factor in the USA. In conclusion, we are not underfitting or overfitting; we are under-informing. No matter how much we tune the model, residuals will persist if the features do not reflect the hidden drivers of LOS. This is a data limitation, not purely a modeling one.
