## Introduction:
This project conducts an exploratory data analysis (EDA) on a curated cohort of spinal cord injury (SCI) patients admitted to the ICU, derived from the MIMIC-IV clinical database. The dataset (ICU_SPINAL_PATIENT_FEATURES_VITALS_EXT) consolidates clinical, demographic, diagnostic, and administrative variables using SQL-based feature engineering. The objective of this EDA is to uncover trends, anomalies, and potential relationships that influence ICU length of stay (LOS), while assessing data quality and distribution characteristics.

Through visualizations and statistical profiling, the analysis explores variables such as dx_count, in_hospital_mortality, discharge_location, admission_type, insurance, ethnicity, and detailed injury types. This foundational step identifies data imbalances (e.g., class skews in categorical variables), non-Gaussian distributions (e.g., LOS), outliers (extreme LOS values), and potential confounding factors (e.g., mortality truncation). Insights from this EDA serve to inform future modeling steps and refine feature preprocessing strategies.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

## Exploratory Data Analysis

```python
icu_df = pd.read_csv("ICU_SPINAL_PATIENT_FEATURES_VITALS_EXT.csv")
icu_df.head(10)
```

## Distribution on Stay Time

```python

# Display the columns to check the time-related ones
print("Columns in the dataset:")
print(icu_df.columns)

# Convert the in time and out time columns to datetime objects
# Replace 'in_time' and 'out_time' with your actual column names
icu_df['intime'] = pd.to_datetime(icu_df['intime'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
icu_df['outtime'] = pd.to_datetime(icu_df['outtime'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Check for any conversion issues
print("\nExample Time Conversion:")
print(icu_df[['intime', 'outtime']].head())

# Calculate the stay time in hours (you could also compute days by dividing by 24)
icu_df['stay_time_hours'] = (icu_df['outtime'] - icu_df['intime']).dt.total_seconds() / 3600.0

# Check the computed stay time
print("\nStay Time (in hours):")
print(icu_df['stay_time_hours'].describe())

# Plot the distribution of stay time
plt.figure(figsize=(10, 6))
sns.histplot(icu_df['stay_time_hours'], kde=True)
plt.title('Distribution of Stay Time (Hours)')
plt.xlabel('Stay Time (Hours)')
plt.ylabel('Frequency')
plt.show()
```

## Distribution Insight

Looking at the pattern distribution of stay time among all patients in the dataset, we see that the hours are mostly between 0 to about 250 hours with some near the 500 mark. Outliers can be seen beyond this point causing the overall distribution to be skewed to the right. The length of stay graph exhibits positive skewness. Furthermore, there is a long tail of high individual LOS values that reach over 2,000 hours. This skewness is further reflected in the standard deviation (144.29). Additionally, the mean (99.88) is much higher than the median (52.20). This means that outliers can affect our predictive model, and tree-based models are likely going to be our best option because of non-normality and outliers.    

## Top 10 Injury Type with Stay Time

```python
if 'long_title' in icu_df.columns:
    # Identify the top 10 injury types by count
    top10_injury_types = icu_df['long_title'].value_counts().head(10).index.tolist()
    
    # Filter the dataframe to include only these top 10 injury types
    filtered_df = icu_df[icu_df['long_title'].isin(top10_injury_types)]
    
    # Create the box plot, ensuring the x-axis shows the top 10 injury types in order
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='long_title', y='stay_time_hours', data=filtered_df, order=top10_injury_types)
    plt.title('Stay Time (Hours) by Top 10 Injury Types')
    plt.xlabel('Injury Type')
    plt.xticks(rotation=90)
    plt.ylim(0, 700)
    plt.ylabel('Stay Time (Hours)')
    plt.show()
else:
    print("Column 'long_title' not found. Replace it with an appropriate column name.")
```

## Top 10 Insight

For the top 10 injury types, the median LOS ranges between 50 to 120 hours (~2 to 5 days). Furthermore, there are many cases where LOS exceeds 200 hours, and some even reach 600 to 700 hours. Injuries such as spinal stenosis, lumbar region, and cervical region show wide IQR and more outliers. Injuries, such as Secondary malignant neoplasm of the brain and spinal cord, show a tighter IQR. Clinically speaking, injuries involving hardware complications may require more extensive ICU monitoring, which may help explain the longer and more variable LOS.

The box plots for the button 10 injury types are almost non-existent, indicating how these injury types have very few patients and may indicate rare injuries. ## Age with Stay Time

```python
if 'anchor_age' in icu_df.columns:
    # Define the bins and corresponding labels
    bins = [0, 20, 40, 60, 80, 100]  # Adjust these bins as needed
    labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
    
    # Create a temporary series for age bins without modifying icu_df
    age_bins = pd.cut(icu_df['anchor_age'], bins=bins, labels=labels, include_lowest=True)
    
    # Option 1: Plot directly using the temporary series
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=age_bins, y=icu_df['stay_time_hours'])
    plt.ylim(0, 400)  # Adjust the y-axis limits as needed
    plt.title('Age Bins vs. Stay Time (Hours)')
    plt.xlabel('Age Bins')
    plt.ylabel('Stay Time (Hours)')
    plt.show()

else:
    print("Column 'anchor_age' not found. Please check your dataset.")
```

## Age Insight

In the plot, we see that stay time decreases when the patient's age increases. Although it is typically seen at the opposite since younger people tend to be more healthy and have quicker recovery times, this does not consider possible death which stay time tracking.

```python

if 'dod' in icu_df.columns and 'anchor_age' in icu_df.columns:
    # Convert 'dod' to datetime (errors become NaT)
    icu_df['dod'] = pd.to_datetime(icu_df['dod'], errors='coerce')
    
    # Filter rows with a valid date of death
    deaths_df = icu_df.dropna(subset=['dod'])
    
    # Group by anchor_age and count the number of deaths
    deaths_count = deaths_df.groupby('anchor_age').size().reset_index(name='num_deaths')
    
    # Option 1: Bar plot using seaborn
    plt.figure(figsize=(12, 6))
    sns.barplot(x='anchor_age', y='num_deaths', data=deaths_count, palette='viridis')
    plt.xlabel('Anchor Age')
    plt.ylabel('Number of Deaths')
    plt.title('Number of Deaths by Anchor Age')
    plt.xticks(rotation=90)
    plt.show()
```

## Number of Deaths Insight

For the mortality bar plot, mortality increases steadily with age, especially after 60, where the slope becomes steeper, indicating age-related vulnerability. Mortality peaked sharply at age 91 (ages >89 are all listed as 91 in the MIMIC database). Our boxplot shows an overall decrease in LOS, variability, and max values as age increases. Combining both graphs, the observed lower length of stay (LOS) in older spinal cord injury patients is likely influenced by their significantly higher mortality rates, as evidenced by the steep increase in deaths with age in your dataset. Older patients often experience early ICU mortality due to greater frailty and comorbidities, which artificially truncates their LOS and skews the distribution downward. This creates a statistical bias where shorter stays do not reflect faster recovery but rather early death, negatively influencing predictive modeling.

## Correlation Matrix

```python
# Select only numeric columns for the correlation matrix
numeric_cols = icu_df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = icu_df[numeric_cols].corr()

# Print correlations with stay_time_hours sorted in descending order
if 'stay_time_hours' in corr_matrix.columns:
    corr_stay_time = corr_matrix['stay_time_hours'].sort_values(ascending=False)
    print("Correlation of features with stay_time_hours:")
    print(corr_stay_time)
else:
    print("Column 'stay_time_hours' not found in the numeric data.")

# Plot the full correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

## Correlation Matrix Insight

Our correlation matrix shows no multicollinearity, which is a positive. Furthermore, lab results, vital signs, and diagnosis count demonstrate some correlation and predictive power; however, several negatively correlated columns, such as 'MIN_SP02', signify early mortality and the need to adjust for a more accurate model.

## Conclusion (Based on EDA Section Only):

The exploratory analysis highlights several clinically relevant patterns in the ICU SCI cohort. Length of stay (LOS) exhibits a highly skewed distribution with extreme outliers, necessitating careful preprocessing in downstream modeling. Categorical features such as insurance, ethnicity, and discharge_location display imbalances and rare categories, which could introduce noise if not addressed. Features like dx_count, admission_type, and in_hospital_mortality show visible associations with LOS and may serve as strong predictors.

Additionally, early mortality cases are associated with shorter LOS, which introduces censoring effects that must be accounted for to avoid misleading inferences. The presence of multicollinearity among administrative and diagnostic features also emerged, suggesting the need for dimensionality reduction or hierarchical feature grouping. Overall, the EDA provides critical guidance for refining the dataset structure and selecting relevant features, forming a data-driven basis for any subsequent predictive modeling.
